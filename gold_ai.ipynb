```python

# Import model structure (Sequential = layers stacked one after another)
model = Sequential()

# -----------------------------
# First LSTM Layer
# -----------------------------

# LSTM(units=50):
# → 50 neurons (memory cells) that learn patterns from time-series data

# return_sequences=True:
# → VERY IMPORTANT
# → This tells LSTM to pass the full sequence to the next LSTM layer
# → Required because we are stacking another LSTM layer after this

# input_shape=(X_train.shape[1], X_train.shape[2]):
# → Defines input format:
#    X_train.shape[1] = number of time steps (e.g., 60 days)
#    X_train.shape[2] = number of features (e.g., 2 → Gold + USD/INR)

model.add(LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))

# Dropout(0.2):
# → Prevents overfitting (model memorizing instead of learning)
# → Randomly drops 20% of neurons during training
# → Helps model generalize better

model.add(Dropout(0.2))

# -----------------------------
# Second LSTM Layer
# -----------------------------

# LSTM(50):
# → Another layer to learn deeper patterns
# → return_sequences=False (default)
# → Means this layer outputs only final result (not full sequence)

model.add(LSTM(50))

# Again dropout for regularization
model.add(Dropout(0.2))

# -----------------------------
# Output Layer
# -----------------------------

# Dense(1):
# → Fully connected layer
# → Outputs a single value → predicted gold price (INR)

model.add(Dense(1))
```

```python
initial_balance = 100000
balance = initial_balance

position = None
buy_price = 0
quantity = 0

equity_curve = []

transaction_cost = 0.001  # 0.1%
stop_loss = 0.02  # 2%

for i in range(len(signals)):

    price = actual_prices[i]

    # STOP LOSS
    if position == "BUY":
        if price < buy_price * (1 - stop_loss):
            balance = quantity * price
            quantity = 0
            position = None

    # BUY (use full capital)
    if signals[i] == "BUY" and position is None:
        quantity = balance / price
        balance = 0
        buy_price = price
        position = "BUY"

    # SELL
    elif signals[i] == "SELL" and position == "BUY":
        balance = quantity * price
        balance -= balance * transaction_cost  # brokerage
        quantity = 0
        position = None

    # Equity tracking
    if position == "BUY":
        equity = quantity * price
    else:
        equity = balance

    equity_curve.append(equity)
```
